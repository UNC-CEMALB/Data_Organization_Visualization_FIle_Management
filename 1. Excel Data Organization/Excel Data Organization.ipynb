# Excel Data Organization

*These steps are by no means exhaustive! Rather use this dataset as an example of a way to clean the data. Datasets often come in many different formats from our wet bench colleagues, therefore some steps might need to be added, removed, or amended.*



1.	Prior to **ANY** edits, click “File” --> “Save As” to save a new version file that will be the cleaned version of the data.
    - This is very important in the instance that the original version needs to be referred back to (ie. if data is accidently deleted). 
    
2.	The file needs to be named something indicative of the data it contains followed by the current date (ie. "Allostatic Mediator Data_061622"). 
    - The title should be succinct and descriptive. 
    - It's okay to use dashes or underscores in the name of the title, but stay away from special characters.

<img src="2. Excel Data Files/Picture1.png"  />

3.	In Excel, remove extraneous white space (ie. these lines of empty cells need to be removed). 
    - Excel Trick: Select all lines that need to be removed and press Ctrl + - (minus key) for Mac users. 

<img src="2. Excel Data Files/Picture2.png" />


4.	Missing data (empty cells) should be replaced with “NA”.

5.	Create a tab for metadata that explains what each column represents in the dataset. The metadata tab should be moved to the first tab and entitled “XXXXX_METADATA” (ie. “Allostatic_METADATA"). Relabel the data tab as “XXXX_DATA” (ie. “Allostatic_DATA). 
<img src="2. Excel Data Files/Picture3.png" />

6.	Abbreviate and capitalize categorical data. Add those abbreviations to the metadata. 
    - This helps reducing typing and potential typing mistakes in R.
<img src="2. Excel Data Files/Picture4.png" />


7.  Alphabetize (sort) the data by the categorical variable of interest (ie. Group for this dataset).
    - This organizes the data and sets it up for the next step. 
8.	Relabel the subject number/identifier column as “Original_Subject_Number” and create an ordinal subject number column labeled “Subject_Number”. 
    - 	R reads in spaces between words as periods, therefore it’s common practice to replace spaces with underscores when doing data analysis in R. 
    - Avoid using dashes in column names or anywhere else in the dataset. 
<img src="2. Excel Data Files/Picture5.png" />


9.	Remove dashes from the data and replace Greek letters with first letter of the word in English (ie. TNF-$\alpha$ becomes TNFa). 
10.	Bold all column names and center all data. 
<img src="2. Excel Data Files/Picture6.png" />

11.	Create a “Subject_ID” column that combines the subject number with the variable of interest (ie. Group). 
    - This is useful for analyses to identify outliers by the subject number and the group. 
<img src="2. Excel Data Files/Picture7.png" />
12.	 Convert the data from a wide to long format if necessary and categorize the variable types.
